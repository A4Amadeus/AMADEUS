```python
import matplotlib.pyplot as plt
import numpy as np
f=lambda x: np.cos(x)*x**2-x
def derivative(f,x):
  delta_x=1/1000000
  return (f(x+delta_x)-f(x))/delta_x
def find_local_min(start_point,step,num_itr):
  x=start_point
  x_history=[x]
  for i in range(num_itr):
    grad_x=derivative(f,x)
    x=x-step*grad_x
    x_history.append(x)
    new_grad_x=derivative(f,x)
    if new_grad_x==0 or grad_x*new_grad_x<0:
      print("local minmum point at x=",x,"y=",f(x))
      break
  return x_history
x_history=find_local_min(start_point=0,step=0.01,num_itr=300)
x_range=np.linspace(-5,5,1000)
plt.plot(x_range,f(x_range),label="f(x)")
plt.axvline(-3.5,linestyle='--',color='k')
plt.axvline(3.7,linestyle='--',color='r')
plt.plot(x_history,f(np.array(x_history)),'rx',label='update x')
plt.legend(loc="best")
plt.show()
```
